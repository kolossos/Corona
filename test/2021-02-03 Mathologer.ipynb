```python
%matplotlib inline
#from numpy import arange, cos, linspace, pi, sin, random
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

plt.rcParams['figure.figsize']=(7,7)
%config InlineBackend.figure_format="jpg"
```

```python
n=200
points=np.arange(n)
```

```python
x=np.cos(2*np.pi*points/n)
y=np.sin(2*np.pi*points/n)
p=np.array([x,y]).T
```

```python
i=34
plt.plot(x,y);
for j in np.arange(n):
    plt.plot([p[j,0],p[int(i*j)%n,0]],[p[j,1],p[int(i*j)%n,1]]);
```

```python
def f(i):
    xxx=[p[0,0],p[int(i*0)%n,0]]
    yyy=[p[0,1],p[int(i*0)%n,1]]

    for j in np.arange(n)[1:]:
        xxx= np.vstack([xxx,[p[j,0],p[int(i*j)%n,0]]])
        yyy= np.vstack([yyy,[p[j,1],p[int(i*j)%n,1]]])

    plt.plot(xxx.T,yyy.T, '-');
```

```python
interact(f, i=(0,100.101));
```
